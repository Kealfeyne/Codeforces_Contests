```python
def get_max_sum(lst):
  result = lst[0]
  sum = 0
  for i in range(len(lst)):
    sum += lst[i]
    result = max(result, sum)
    sum = max(sum, 0)
  return result
```

```python
def process_list(n, k, main_list):
  second_list = [0] * len(main_list)
  for i in range(2**n - 2**k + 1):
    if second_list[i] == 0:
      main_list[i], main_list[i + 2**k] = main_list[i + 2**k], main_list[i]
      second_list[i] = 1
      second_list[i + 2**k] = 1
  result = get_max_sum(main_list)
  return main_list, result
```

```python
input_n = int(input())
editing_list = list(map(int, input().split(' ')))
input_q = int(input())

for index in range(input_q):
  input_k = int(input())
  editing_list, max_sum = process_list(input_n, input_k, editing_list)
  print(max_sum)
```
